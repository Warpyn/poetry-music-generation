```python
def plot_emotions(kept_emotions, emotion_colors = emotion_color2):
    emotions = [emotion['label'] for emotion in kept_emotions]
    probabilities = [emotion['score'] for emotion in kept_emotions]
    
    total_prob = sum(probabilities)

    if total_prob < 1.0:
        other_emotions_prob = 1.0 - total_prob
        emotions.append("other emotions")
        probabilities.append(other_emotions_prob)
    
    pie_colors = [emotion_colors[e] for e in emotions]

    plt.figure(figsize=(5, 5))  # Set the figure size
    plt.pie(probabilities, labels=emotions, autopct='%1.1f%%', colors=pie_colors, startangle=140)
    plt.axis('equal')

    plt.show()
    
```

```python
def wrapper(file_path = "./testData/shortPoemsForTesting.csv"):
    all_original_emotions, all_VA, all_kept_emotions = compare_emotions(file_path)
    #print(all_kept_emotions)
    print(all_VA)
    for emos in all_kept_emotions:
        #print(emos)
        plot_emotions(emos)

    return
```

```python
def plot_VA(poem_data_path = "./testData/poemWithEmotions.csv"):
    df = pd.read_csv(poem_data_path)
    df = df.head(30)
    

    valences = []
    arousals = []
    poems = []
    emotions = []
    colors = []
    poem_count = 0
    for _, row in df.iterrows():
        poem = row['Poem'] 
        emotion = row['Emotion']

        if len(poem) > 2000:
            continue

        my_poem = poem_to_VA(poem)
        valence, arousal = my_poem.get_VA_per_sentence()[0]
        valences.append(valence)
        arousals.append(arousal)
        poems.append(poem)
        emotions.append(emotion)
        colors.append(emotion_colors[emotion])
        poem_count+=1
    
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.scatter(valences, arousals, c=colors, alpha=0.5)

    # ax.axhline(0, color='black', linewidth=0.8)
    # ax.axvline(0, color='black', linewidth=0.8)

    ax.set_xlim(-1.25, 1.25)
    ax.set_ylim(-1.25, 1.25)

    centerx = 0
    centery = 0

    ax.spines['left'].set_position(('data', centerx))
    ax.spines['bottom'].set_position(('data', centery))
    ax.spines['right'].set_position(('data', centerx - 1))
    ax.spines['top'].set_position(('data', centery - 1))

    for side in ['right', 'top']:
        ax.spines[side].set_color('none')

    ax.xaxis.set_label_coords(0.5, -0.025)
    ax.yaxis.set_label_coords(-0.025, 0.5)

    ax.set_xlabel('Valence')
    ax.set_ylabel('Arousal')
    ax.set_title('Valence and Arousal in Poems by Emotion')

    #ax.legend(emotion_colors.keys(), loc='upper right', bbox_to_anchor=(1.15, 1))

    plt.show()

    poems_data = pd.DataFrame({
        'Poem': poems,  
        'Valence': valences,
        'Arousal': arousals,
        'Emotion': emotions  
    })

    output_csv_path = 'testOutput.csv'  
    poems_data.to_csv(output_csv_path, index=False)

    return poem_count
```
